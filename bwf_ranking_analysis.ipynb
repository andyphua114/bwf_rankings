```python
import pandas as pd
import numpy as np
import re
```

```python
metadata_df = pd.read_csv('filename_url.csv')
```

```python
metadata_df.tail()
```

```python
metadata_tuple = [tuple(r) for r in metadata_df.to_numpy()]
```

```python
final = pd.read_excel("rankings\\" + metadata_tuple[0][0], sheet_name = 0, header = 3)
cols = ['BWF ID', 'Last name', 'First name', 'Ranking']
final['First name'] = final['First name'].str.lower()
final['Last name'] = final['Last name'].str.lower()
final = final[cols]
final = final.rename(columns = {'Ranking': metadata_tuple[0][0].replace(' ', '')[2:12]})

for md in metadata_tuple[1:]:
    
    hold = pd.read_excel("rankings\\" + md[0], sheet_name = 0, header = 3)
    
    hold['First name'] = hold['First name'].str.lower()
    hold['Last name'] = hold['Last name'].str.lower()
    hold = hold[cols]
    hold = hold.rename(columns = {'Ranking': md[0].replace(' ', '')[2:12]})
    
    final = final.merge(hold, how = 'outer', on = ['BWF ID', 'Last name', 'First name'])
    print(md[0])
```

```python
final.head()
```

```python
ranking = final.groupby('BWF ID', as_index = False).max(numeric_only=True)
```

```python
ranking
```

```python
id_toname = final.groupby('BWF ID', as_index = False).agg({'First name': 'last', 'Last name' : 'last'})
```

```python
id_toname[id_toname['BWF ID'] == 5692]
```

```python
final_ranking = id_toname.merge(ranking, on = 'BWF ID', how = 'outer')
```

```python
final_ranking
```

```python
alpha_cols = ['BWF ID', 'First name', 'Last name']
all_cols = list(final_ranking.columns)
rank_cols = list(set(all_cols) - set(alpha_cols))
```

```python
final_ranking.to_csv('historical_bwf_ranking.csv', index = False)
```
